```python
#Bernoulli


import numpy as np            
from scipy.stats import binom
import matplotlib.pyplot as plt
import math

def fbern(p,a,n):
    return ( (p/a)**a *( (p-1)/(a-1) )**(1-a) )**n


def bio(p,a,n):
    return 1- binom.cdf(a*n,n,p)

def lg(n,p):
    return 1/n*math.log(p)

def compare(p,a,n):
    app=fbern(p,a,n)
    real=bio(p,a,n)
    return app,real


#compare(0.5,0.3,100)


#when p=0.5, a=0.8
n=[0]*40
for i in range(40):
    n[i]=5*(i+1)
prob=[0]*40                                     #real probability
aprox=[0]*40                                    #large deviation
log_n=[0]*40                                    #1/n * log(P)
ratio=[0]*40

for i in range(40):
    aprox[i]=fbern(0.5,0.8,n[i])
    prob[i]=bio(0.5,0.8,n[i])
    if(prob[i]!=0):
        ratio[i]=aprox[i]/prob[i]
        log_n[i]=1/n[i] * math.log(prob[i])
    
    

print('real probability:')
print(prob)
print('approximation:')
print(aprox)
print('ratio:')
print(ratio[0:36])
print('1/n*lg(p)')
print(log_n[0:36])


plt.plot(n[0:36], prob[0:36], "k-")
plt.show()

plt.plot(n, aprox, "y")
plt.show()

plt.plot(n[0:30], prob[0:30], "k-", label = "real probability")
plt.plot(n[0:30], aprox[0:30], "y", label = "the approximate probability")
plt.xlabel(r"$n$", fontsize = 10)                 #let the label of x be year.
plt.ylabel(r"$p$", fontsize = 10)            #let the label of y be emissions.
plt.legend(loc = "upper right", fontsize = 12.5)        #name the lines in the questions
plt.show()

plt.plot(n[0:30], log_n[0:30], "k")
plt.xlabel(r"$n$", fontsize = 10)
plt.ylabel(r"$plog(P)$", fontsize = 10)
plt.show()

#note: there is numerical error for binom.cdf when n > 150
#      theoretically, should be tend to -0.192744 when a=0.8, p=0.5
#      relationship between ratio and n is sqrt(n)
```

```python
#Geometric

import numpy as np
import math
from scipy.stats import nbinom
import matplotlib.pyplot as plt

def fge(p,a,n):
    return  ( p*(1+a) * ( a/((1-p)*(1+a)) )**(-a) )**n

def geo(p,a,n):
    return 1- nbinom.cdf(a*n, n, p)

def lg(n,p):
    return 1/n*math.log(p)

def Fenchel(p,a):
    return a*math.log(a/((1-p)*(1+a)) )-math.log(p*(1+a))


#when p=0.4, a=3               as the expected value is 1/0.4=2.5
n=[0]*20
for i in range(20):
    n[i]=10*(i+1)
prob=[0]*20                                     #real probability
aprox=[0]*20                                    #large deviation
log_n=[0]*20                                    #1/n * log(P)
ratio=[0]*20

for i in range(20):
    aprox[i]=fge(0.4,3,n[i])
    prob[i]=geo(0.4,3,n[i])
    if(prob[i]!=0):
        ratio[i]=aprox[i]/prob[i]
        log_n[i]=1/n[i] * math.log(prob[i])


print('real probability:')
print(prob[0:15])
print('approximation:')
print(aprox)
print('ratio:')
print(ratio[0:15])
print('the fenchel-legenfre value is %f',Fenchel(0.4,3))
print('1/n*lg(p)')
print(log_n[0:15])

    
    
plt.plot(n[0:15], prob[0:15], "k-")
plt.show()

plt.plot(n, aprox, "y")
plt.show()

plt.plot(n[0:15], prob[0:15], "k-")
plt.plot(n[0:15], aprox[0:15], "y")
plt.show()

plt.plot(n[0:15], log_n[0:15], "k")
plt.show()




#when p=0.4, a=2.6
n=[0]*20
for i in range(20):
    n[i]=10*(i+1)
prob=[0]*20                                     #real probability
aprox=[0]*20                                    #large deviation
log_n=[0]*20                                    #1/n * log(P)
ratio=[0]*20

for i in range(20):
    aprox[i]=fge(0.4,2.6,n[i])
    prob[i]=geo(0.4,2.6,n[i])
    if(prob[i]!=0):
        ratio[i]=aprox[i]/prob[i]
        log_n[i]=1/n[i] * math.log(prob[i])


print('real probability:')
print(prob)
print('approximation:')
print(aprox)
print('ratio:')
print(ratio)
print('the fenchel-legenfre value is %f',Fenchel(0.4,2.6))
print('1/n*lg(p)')
print(log_n)

    
    
plt.plot(n, prob, "k-")
plt.show()

plt.plot(n, aprox, "y")
plt.show()

plt.plot(n, prob, "k-", label = "real probability")
plt.plot(n, aprox, "y", label = "the approximate probability")
plt.xlabel(r"$n$", fontsize = 10)                 #let the label of x be year.
plt.ylabel(r"$p$", fontsize = 10)            #let the label of y be emissions.
plt.legend(loc = "upper right", fontsize = 12.5)        #name the lines in the questions
plt.show()

plt.plot(n, log_n, "k")
plt.xlabel(r"$n$", fontsize = 10)
plt.ylabel(r"$\frac{1}{n} log(P)$", fontsize = 10)
plt.show()


#note: there is numerical error for nbinom.cdf when a=3 n > 150 
#      seems like the approximation is rely bad
#      theoretically, should be tend to -0.216932 when p=0.4, a=3
#      theoretically, should be tend to -0.164285 when p=0.4, a=2.6
```

```python
# Normal distribution(0,1)

import numpy as np
import math
from scipy.stats import norm
import matplotlib.pyplot as plt

def fno(mean,var,a,n):
    return  ( math.exp(-0.5*(a-mean)**2/var ) )**n

def nor(mean,var,a,n):
    return 1- norm.cdf(a*n,n*mean,(n*var)**0.5 )

def lg(n,p):
    return 1/n*math.log(p)

def Fenchel(mean,var,a):
    return 0.5*(a-mean)**2 /var

#when mean=0, var=1, a=0.1
n=[0]*30
for i in range(30):
    n[i]=10*(i+1)
prob=[0]*30                                     #real probability
aprox=[0]*30                                    #large deviation
log_n=[0]*30                                    #1/n * log(P)
ratio=[0]*30


for i in range(30):
    aprox[i]=fno(0,1,0.1,n[i])
    prob[i]=nor(0,1,0.1,n[i])
    if(prob[i]!=0):
        ratio[i]=aprox[i]/prob[i]
        log_n[i]=1/n[i] * math.log(prob[i])


print('real probability:')
print(prob)
print('approximation:')
print(aprox)
print('ratio:')
print(ratio)
print('the fenchel-legenfre value is %f',Fenchel(0,1,0.1))
print('1/n*lg(p)')
print(log_n)


plt.plot(n, prob, "k-")
plt.show()

plt.plot(n, aprox, "y")
plt.show()

plt.plot(n, prob, "k-")
plt.plot(n, aprox, "y")
plt.show()

plt.plot(n, log_n, "k")
plt.show()

#when mean=0, var=1, a=0.2
n=[0]*30
for i in range(30):
    n[i]=10*(i+1)
prob=[0]*30                                     #real probability
aprox=[0]*30                                    #large deviation
log_n=[0]*30                                    #1/n * log(P)
ratio=[0]*30


for i in range(30):
    aprox[i]=fno(0,1,0.2,n[i])
    prob[i]=nor(0,1,0.2,n[i])
    if(prob[i]!=0):
        ratio[i]=aprox[i]/prob[i]
        log_n[i]=1/n[i] * math.log(prob[i])


print('the fenchel-legenfre value is %f',Fenchel(0,1,0.2))

print('real probability:')
print(prob)
print('approximation:')
print(aprox)
print('ratio:')
print(ratio)
print('1/n*lg(p)')
print(log_n)


plt.plot(n, prob, "k-")
plt.show()

plt.plot(n, aprox, "y")
plt.show()

plt.plot(n, prob, "k-", label = "real probability")
plt.plot(n, aprox, "y", label = "the approximate probability")
plt.xlabel(r"$n$", fontsize = 10)                 #let the label of x be year.
plt.ylabel(r"$p$", fontsize = 10)            #let the label of y be emissions.
plt.legend(loc = "upper right", fontsize = 12.5)        #name the lines in the questions
plt.show()

plt.plot(n, log_n, "k")
plt.xlabel(r"$n$", fontsize = 10)
plt.ylabel(r"$\frac{1}{n} log(P)$", fontsize = 10)
plt.show()



#note: there is numerical error for binom.cdf when n > 300
#      theoretically, should be tend to 0.00500 when mean=0, var=1, a=0.1

```

```python
# uniform on[0,1]

from scipy.integrate import quad
import numpy as np
import math
from scipy.stats import gamma
import matplotlib.pyplot as plt
import sympy

#first we find the fenchel-legenfre value by Newton's method:

def F(t,a):
    return a-math.exp(t)/(math.exp(t)-1)+1/t
    
def Fp(t):      #Fp(x) is the derivative of f
    return -1/t**2 - math.exp(t)/(math.exp(t)-1)**2
    
def G(x,a):       #G(x) is used to evaluate the iteration function
    x1=F(x,a)
    x2=Fp(x)
    return x - (x1/x2)


def Newton(x,acc,a):
    a=0.6
    x_old = np.inf
    x_n = x
    while abs(x_n - x_old) > acc:
        x_old = x_n
        x_n = G(x_n,a)
    return x_n

#set a=0.6, and we find the solution near 0.8 with accuracy 0.0001

t0=Newton(0.8,0.0001,0.6)          #with accuracy 0.0001
print("t0 is: ")
print(t0)


def Fenchel(a):     #on interval [0,1]
    return a*t0-math.log(math.exp(t0)-1)+math.log(t0)


def funiform(a,n):
    return (math.exp(-Fenchel(a)) )**n


def nCr(n,r):
    f = math.factorial
    return f(n) / f(r) / f(n-r)

def pdfuni(x,n):          #the pdf of Sn of the uniform distribution

    if x<=n:
        integer=int(x)
        sum=0
        for k in range(integer+1):        
            sum += ( (-1)**k )*(x-k)**(n-1) * nCr(n,k)
        return sum/math.factorial(n-1)
    else:
        return 0


def cdfuni(x,n):
    sum=0
    integer=int(x)
    for k in range(integer+1):        
        sum += ( (-1)**k )*(x-k)**(n) * nCr(n,k)
    return sum/math.factorial(n)



def uni(a,n):
    return 1 - cdfuni(a*n,n)


def lg(n,p):
    return 1/n*math.log(p)

n=[0]*60
for i in range(60):
    n[i]=i+1
prob=[0]*60                                     #real probability
aprox=[0]*60                                    #large deviation
log_n=[0]*60                                    #1/n * log(P)
ratio=[0]*60


for i in range(60):
    aprox[i]=funiform(0.6,n[i])
    prob[i]=uni(0.6,n[i])
    if(prob[i]>0):
        ratio[i]=aprox[i]/prob[i]
        log_n[i]= 1/n[i] * math.log(prob[i])



print('real probability:')
print(prob)
print('approximation:')
print(aprox)
print('ratio:')
print(ratio[0:51])
print('the fenchel-legenfre value is %f',Fenchel(0.6))
print('1/n*lg(p)')
print(log_n[0:51])


plt.plot(n[0:51], prob[0:51], "k-")
plt.show()

plt.plot(n, aprox, "y")
plt.show()

plt.plot(n[0:51], prob[0:51], "k-", label = "real probability")
plt.plot(n[0:51], aprox[0:51], "y", label = "the approximate probability")
plt.xlabel(r"$n$", fontsize = 10)                 #let the label of x be year.
plt.ylabel(r"$p$", fontsize = 10)            #let the label of y be emissions.
plt.legend(loc = "upper right", fontsize = 12.5)        #name the lines in the questions
plt.show()

plt.plot(n[0:51], log_n[0:51], "k")
plt.xlabel(r"$n$", fontsize = 10)
plt.ylabel(r"$\frac{1}{n} log(P)$", fontsize = 10)
plt.show()

#numerical error occur when n>60
```
